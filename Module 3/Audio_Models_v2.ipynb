### Voice Chat with LLM

```python
from langchain_openai import ChatOpenAI

# to be able to use LLM to generate text for the audio
llm = ChatOpenAI()
```

```python
response = llm.invoke("How do volcanoes work? Reply in one sentence.")
print(response.content)
```

```python
def get_reply_in_voice(text):
  response = llm.invoke(text)
  print(response.content)
  audio = client.generate(
                      text = response.content,
                      voice = "Rachel"
                  )
  play(audio, notebook = True)
```

```python
get_reply_in_voice("Who was the first Nobel prize winner? Give me a one phrase reply")
```

```python
def get_multilingual_reply(text):
  response = llm.invoke(text)
  audio = client.generate(
    text = response.content,
    voice = "Rachel",
    model = "eleven_multilingual_v2"
)

  play(audio, notebook = True)
```

```python
get_multilingual_reply("आप कैसे हैं?")
```

```python
get_multilingual_reply("మీరు ఎలా ఉన్నారు?")
```

### Meeting Summariser using whisper

```python
# Summarise the meeting

def summarise_meeting(meeting_audio):
  audio_file = open(meeting_audio, "rb")
  transcript = client.audio.transcriptions.create(
                        model="whisper-1",
                        file=audio_file,
                        response_format="text"
                      )

  llm = ChatOpenAI(model = "gpt-3.5-turbo")
  result = llm.invoke(f""" Summarise the below transcript into 3 bullet points
                  {transcript}
  """)

  return result.content

```

```python
print(summarise_meeting("/content/GenAI_Intro_Audio.mp3"))
```
